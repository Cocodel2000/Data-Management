```python
from bs4 import BeautifulSoup
import requests

ticker = "AAPL"
url = f"https://finance.yahoo.com/quote/{ticker}"

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# find all table rows in the left summary table
left_table_rows = soup.find('div', {'data-test': 'left-summary-table'}).find_all('tr')

# find all table rows in the right summary table
right_table_rows = soup.find('div', {'data-test': 'right-summary-table'}).find_all('tr')

# create a dictionary to store the extracted data
data = {}

# loop through all the left summary table rows and extract the data
for row in left_table_rows:
    cols = row.find_all('td')
    data[cols[0].text.strip()] = cols[1].text.strip()

# loop through all the right summary table rows and extract the data
for row in right_table_rows:
    cols = row.find_all('td')
    data[cols[0].text.strip()] = cols[1].text.strip()

# print the extracted data
print(data)
```

```python
import requests
from bs4 import BeautifulSoup

ticker = "aapl"
url = f"https://www.tipranks.com/stocks/{ticker}/forecast"

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# find all the rows in the table with the class 'rt-tr-group'
rows = soup.find_all('div', {'class': 'rt-tr-group'})

# loop through each row
for row in rows:
    # extract the analyst's name
    analyst_name = row.find('div', {'class':"colorlink hoverDecorationunderline alignstart truncate fontWeightsemibold mt2"}).text.strip()
    print(f"Analyst's name: {analyst_name}")

    # extract the analyst's company
    analyst_company = row.find('div', {'class': 'truncate'}).text.strip()
    print(f"Analyst's company: {analyst_company}")

    # extract the analyst's price target
    analyst_price_target = row.find('div', {'class': 'fontWeightsemibold'}).text.strip()
    print(f"Analyst's price target: {analyst_price_target}")

    # extract the analyst's rating
    analyst_rating = row.find('div', {'class': "fontWeightbold  colorpale  textTransformuppercase"}).text.strip()
    print(f"Analyst's rating: {analyst_rating}")

    # extract the upside potential
    upside_potential = row.find('div', {'class':"flexr_b w12    displayflex colorpale fontWeightsemibold"}).text.strip()
    print(f"Upside potential: {upside_potential}")

    # extract the action performed
    action_performed = row.find('div', {'class': 'action'}).text.strip()
    print(f"Action performed: {action_performed}")

    # extract the time of the forecast
    time_of_forecast = row.find('div', {'class': 'date'}).text.strip()
    print(f"Time of the forecast: {time_of_forecast}")

    # extract the follow button
    follow_button = row.find('div', {'class': 'follow'}).text.strip()
    print(f"Follow button: {follow_button}")

    # extract the news button
    news_button = row.find('div', {'class': 'news'}).text.strip()
    print(f"News button: {news_button}")

    print('\n')
```

```python
#need to be continuated
import requests
from bs4 import BeautifulSoup

aapl = 'aapl'
url = f"https://www.tipranks.com/stocks/{aapl}/forecast"

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# find the table containing the analyst recommendations
table = soup.find('div', {'class': 'analyst-forecast-table'})

# check if there are any recommendations for the given ticker
if not table:
    print(f"No recommendations found for {aapl.upper()} ticker.")
else:
    # loop through each row of the table
    for row in table.find_all('div', {'class': 'rt-tr-group'})[1:]:
        # extract the analyst's name
        analyst_name = row.find('div', {'class': 'name'}).text.strip()
        
        # extract the analyst's company
        analyst_company = row.find('div', {'class': 'company'}).text.strip()
        
        # extract the analyst's price target
        analyst_price_target = row.find('div', {'class': 'target-price'}).text.strip()
        
        # extract the analyst's rating
        analyst_rating = row.find('div', {'class': 'rating'}).text.strip()
        
        # extract the upside potential
        upside_potential = row.find('div', {'class': 'potential'}).text.strip()
        
        # extract the action performed by the analyst
        action_performed = row.find('div', {'class': 'action'}).text.strip()
        
        # extract the time of the forecast
        time_of_forecast = row.find('div', {'class': 'time'}).text.strip()
        
        # print the extracted information
        print(f"Analyst's name: {analyst_name}")
        print(f"Analyst's company: {analyst_company}")
        print(f"Analyst's price target: {analyst_price_target}")
        print(f"Analyst's rating: {analyst_rating}")
        print(f"Upside potential: {upside_potential}")
        print(f"Action performed: {action_performed}")
        print(f"Time of the forecast: {time_of_forecast}")
        print('\n')
```
